```python
# Setting conditions for chart creation
# Function to create y data (report)

def get_datasets(y, labels): #List by List
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label': labels[i],
                'data': y[i]
            })
        return datasets
    else: #List one value
        return [
            {
                'label': labels[0],
                'data': y
            }
        ]
```

```python
# Create chart title 
def set_title(title=''):
    if title != '':
        display = 'true'
    else:
        display = 'false'
    return {
        'title': title,
        'display': display
    }
```

```python
# Function that creates the graph (Dictionary representative of the data)
def create_chart(x, y, labels, kind = 'bar', title = ''):
    
    datasets = get_datasets(y,labels) # Axis y and labels
    options = set_title(title) # Title
    
    chart = {
        'type': kind,
        'data': {
            'labels': x, # Axis x value
            'datasets': datasets # value
        },
        'options': options 
    }
    return chart
```

```python
# API request function through the created dictionary
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}') 
    return resp.content # Storing binary request value
```

```python
# Function that saves the created image
def save_image(path, content):
    with open(path, 'wb') as image: #Write with binary value wb
        image.write(content)
```

```python
# Importing functions that let you view the saved image
from PIL import Image
from IPython.display import display
```

```python
# Function showing the image
def display_image(path):
    img_pil = Image.open(path)
    display(img_pil)
```

```python
# Create the graph

# Axis y1 and y2
y_data_1 = []
for obs in final_data[1::15]: # 15 by 15 days
    y_data_1.append(obs[CONFIRMADOS])
    
y_data_2 = []
for obs in final_data[1::15]:
    y_data_2.append(obs[RECUPERADOS])
    
# Axis x
labels = ['Confirmados', 'Recuperados']
x = []
for obs in final_data[1::15]:
    x.append(obs[DATA].strftime('%d/%m/%Y')) # Datetime into String date with strftime

# Create chart
chart = create_chart(x, [y_data_1, y_data_2], labels, title='Gráfico Confirmado vs Recuperados')
chart_content = get_api_chart(chart)
save_image('my-first-graph.png', chart_content)
display_image('my-first-graph.png')
```

```python
# Generate create QR Code for direct access to the created graphic
# Importing functions parsing link
from urllib.parse import quote
```

```python
# Function request link
def get_api_qrcode(link):
    text = quote(link) # parsing link to url
    url_base = 'https://quickchart.io/qr'
    resp = r.get(f'{url_base}?text={text}')
    return resp.content
```

```python
# Request QR Code
url_base = 'https://quickchart.io/chart'
link = f'{url_base}?c={str(chart)}'
save_image('qr-code.png',get_api_qrcode(link))
display_image('qr-code.png')
```
