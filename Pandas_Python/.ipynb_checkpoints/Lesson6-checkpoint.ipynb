# Renaming and Combining

```python
import pandas as pd
reviews = pd.read_csv("winemag-data-130k-v2.csv", index_col=0)
print("Setup complete")
```

```python
# Utilizando a  funÃ§Ã£o rename() para alterar nomes de Ã­ndices e/ou nomes de colunas.
reviews.rename(columns={'points': 'score'})
```

### OBS: rename() permite renomear valores de Ã­ndice ou coluna especificando um parÃ¢metro de palavra-chave de Ã­ndice ou coluna, respectivamente. Ele suporta uma variedade de formatos de entrada, mas geralmente um dicionÃ¡rio Python Ã© o mais conveniente.

```python
reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
```

```python
#  Renomeando os valores de Ã­ndice com set_index(). E, renomeando o Ã­ndice de linhas e colunsa com rename_axis()
reviews.rename_axis("wines", axis='rows').rename_axis("fields", axis='columns')
```

## Combining

```python
# Utilizando os mÃ©todos concat(), join() e merge() para combinar valores do DataFrame. 
```

```python
# concat()
canadian_youtube = pd.read_csv("CAvideos.csv")
british_youtube = pd.read_csv("GBvideos.csv")

pd.concat([canadian_youtube, british_youtube])
```

```python
# utilizando join()
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
```

#### OBS: Os parÃ¢metros lsuffix e rsuffix sÃ£o necessÃ¡rios aqui porque os dados tÃªm os mesmos nomes de coluna nos conjuntos de dados britÃ¢nicos e canadenses.
