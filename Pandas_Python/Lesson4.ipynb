# Grouping and Sorting

```python
import pandas as pd
```

```python
reviews = pd.read_csv("winemag-data-130k-v2.csv", index_col=0)
```

```python
reviews
```

# groupby() criou um grupo de avaliações que atribuía os mesmos valores de pontos aos vinhos fornecidos.

```python
# Usando o metodo .count()
reviews.groupby('points').points.count()
```

```python
# Usando o metodo min()
reviews.groupby('points').price.min()
```

```python
# Usando o metodo apply()
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
# Chamando duas colunas
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
# gerar um resumo estatístico simples do conjunto de dados com o metodo agg()
reviews.groupby(['country']).price.agg([len, min, max])
```

## Multi-indexes

```python
# metodos multiplos indices
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
mi = countries_reviewed.index
type(mi)
```

```python
# metodo multi-indices => reset_index():
countries_reviewed.reset_index()
```

## Sorting

```python
# obtendo os dados na ordem desejada com o metodo sort_values() 
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

```python
countries_reviewed.sort_values(by='len', ascending=False)
```

```python
# classificando por valores de índice, use o método complementar sort_index()
countries_reviewed.sort_index()
```

```python
countries_reviewed.sort_values(by=['country', 'len'])
```

```python
# Crie uma série cujo índice seja a categoria taster_twitter_handle do conjunto de dados e 
# cujos valores contem quantas avaliações cada pessoa escreveu

reviews_written = reviews.groupby('taster_twitter_handle').size()
reviews_written
```

```python
# Crie uma Série cujo índice seja os preços dos vinhos e cujos valores sejam o número máximo de pontos que um 
# vinho de preço tão alto foi dado em uma resenha. 
# Classifique os valores por preço, em ordem crescente (de modo que 4,0 dólares fique no topo 
# e 3300,0 dólares na parte inferior).
best_rating_per_price = reviews.groupby('price')['points'].max().sort_index()
best_rating_per_price
```

```python
# Crie um DataFrame cujo índice seja a categoria de variedade do conjunto de dados e 
# cujos valores sejam os valores mínimo e máximo do mesmo.
price_extremes = reviews.groupby('variety').price.agg([min, max])
price_extremes
```

###  4

```python
# Crie uma variável sorted_varieties contendo uma cópia do dataframe da pergunta anterior
# onde as variedades são classificadas em ordem decrescente com base no preço mínimo
# e depois no preço máximo (para desempate).
sorted_varieties = price_extremes.sort_values(by=['min', 'max'], ascending=False)
sorted_varieties
```

### 5

```python
# Crie uma Série cujo índice seja de revisores e cujos valores sejam a pontuação média de revisão dada por esse revisor. 
reviewer_mean_ratings = reviews.groupby('taster_name').points.mean()
reviewer_mean_ratings
```

```python
# usar o método describe() para ver um resumo do intervalo de valores.
reviewer_mean_ratings.describe()
```

### 6

```python
# Crie uma Série cujo índice seja um MultiIndex de pares {país, variedade}. 
# Por exemplo, um pinot noir produzido nos EUA deve mapear para {"US", "Pinot Noir"}. 
# Classifique os valores na Série em ordem decrescente com base na contagem de vinhos.
country_variety_counts = reviews.groupby(['country', 'variety']).size().sort_values(ascending=False)
country_variety_counts
```
